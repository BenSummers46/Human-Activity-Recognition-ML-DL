# Linear Regression Model

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()

X = np.reshape(duration, (-1, 1))

regr.fit(X, cycling_data['EEtot'])
```

```python
theta0 = regr.intercept_
theta1 = regr.coef_

plt.scatter(X, cycling_data['EEtot'])
plt.plot(X, theta0 + theta1*X)
plt.xlabel('Duration')
plt.ylabel('Total Calories Used')
plt.show()
```

```python
test_duration = [500]

test_duration = np.reshape(test_duration, (-1, 1))

y_pred = regr.predict(test_duration)
print("Calories used with {} seconds of cycling: {:.2f}".format(test_duration[0], y_pred[0]))
```
